```python
import pandas as pd

url="./Phishing.csv"
df=pd.read_csv(url)
```

```python
value_counts=df.Result.value_counts()
df2 = value_counts.rename_axis('Class').reset_index(name='Num_Observations')
df2
```

```python
import matplotlib.pyplot as plt

colors = ["r", "b"]
ax = df2.plot(kind='bar', title ="Occurrences of Phishing", figsize=(15, 10), legend=True, fontsize=12, color=colors)
ax.set_xlabel("Class", fontsize=12)
ax.set_ylabel("Num_Observations", fontsize=12)
plt.show()
```
